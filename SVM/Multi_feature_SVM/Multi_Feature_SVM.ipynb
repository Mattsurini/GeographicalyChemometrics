```python
Data = pd.read_csv('/content/drive/MyDrive/Working Space/Data_coffee/Data_coffee_complete_correlationWithGroup.csv')
Data
```

```python
feature = ['arginine',
 'L',
 'Antioxidant',
 'glycine',
 'B',
 'A',
 'phynylalanine',
 'Apartic_acid',
 'leucine',
 'Serine',
 'alanine',
 'threonine',
 'proline',
 'histidine',
 'caffeic_acid',
 'K',
 'Lactic_acid']
feature
```

```python
Data[feature]
```

```python
X = Data[feature]
y = Data.Groupping
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.33, random_state = 42)
```

```python
linear = svm.SVC(kernel='linear', C=1, decision_function_shape='ovo').fit(X_train, y_train)
rbf = svm.SVC(kernel='rbf', gamma=1, C=1, decision_function_shape='ovo').fit(X_train, y_train)
poly = svm.SVC(kernel='poly', degree=3, C=1, decision_function_shape='ovo').fit(X_train, y_train)
sig = svm.SVC(kernel='sigmoid', C=1, decision_function_shape='ovo').fit(X_train, y_train)
```

```python
linear_pred = linear.predict(X_test)
poly_pred = poly.predict(X_test)
rbf_pred = rbf.predict(X_test)
sig_pred = sig.predict(X_test)
```

```python
accuracy_lin = linear.score(X_test, y_test)
accuracy_poly = poly.score(X_test, y_test)
accuracy_rbf = rbf.score(X_test, y_test)
accuracy_sig = sig.score(X_test, y_test)
print('Linear Kernel:', accuracy_lin)
print('Polynomial Kernel:', accuracy_poly)
print('Radial Basis Kernel:', accuracy_rbf)
print('Sigmoid Kernel:', accuracy_sig)
```
